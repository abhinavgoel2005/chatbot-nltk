```python
import nltk
import numpy as np
import random
import json
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
intents = {
    "intents": [
        {"intent": "greeting", "patterns": ["Hi", "Hello", "Hey", "How are you?", "Good morning"], "responses": ["Hello!", "Hi there!", "Good to see you!", "How can I assist you?"]},
        {"intent": "goodbye", "patterns": ["bye", "Goodbye", "See you later", "Take care"], "responses": ["Goodbye!", "See you later!", "Take care!"]},
        {"intent": "thanks", "patterns": ["Thank you", "Thanks", "I appreciate it"], "responses": ["You're welcome!", "Happy to help!", "Glad I could assist!"]},
        {"intent": "age", "patterns": ["How old are you?", "What's your age?", "How many years have you been alive?"], "responses": ["I'm a bot, I don't age!", "I don't have an age, but I'm always learning!"]},
        {"intent": "name", "patterns": ["What's your name?", "Who are you?", "Tell me your name"], "responses": ["I'm a chatbot!", "I am Chatbot.", "You can call me Bot!"]},
        {"intent": "not_understood", "patterns": ["", "I don't know", "Hmmm", "Uhhh", "Not sure"], "responses": ["Sorry, I didn't quite get that.", "Can you please clarify?", "I didn't understand. Could you ask again?"]},
    ]
}
```

```python
patterns=[]
responses=[]
labels=[]

for intent in intents['intents']:
    for pattern in intent['patterns']:
        patterns.append(pattern)
        responses.append(intent['responses'])
        labels.append(intent['intent'])
```

```python
def preprocess_text(text):
  text=text.lower()
  text=''.join([char for char in text if char not in string.punctuation])
  return text
patterns=[preprocess_text(text) for text in patterns]
```

```python
model=make_pipeline(TfidfVectorizer(),MultinomialNB())
```

```python
x_train=patterns
y_train=labels
model.fit(x_train,y_train)
```

```python
def chatbot_responses(user_input):
  user_input=preprocess_text(user_input)
  predicted_intent=model.predict([user_input])[0]
  for intent in intents["intents"]:
    if intent["intent"]==predicted_intent:
      return random.choice(intent["responses"])
  return random.choice(["soory,anshuL aap itne vo kyu ho","aur haan ritkL aap bhi"])
```

```python
def chat():
  print("hello, I'm the legendary abhinav")
  while True:
    user_input=input("You:")
    if user_input.lower()=='bye':
      print("Chatbot: Goodbye!")
      break
    response=chatbot_responses(user_input)
    print(f"Chatbot:,{response}")
```

```python
chat()
```
